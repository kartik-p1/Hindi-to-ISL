```python
# create syllables of a word
def syllables(word):
        signs = [u'\u0902', u'\u093e', u'\u093f', u'\u0940', 
                 u'\u0941', u'\u0942', u'\u0943', u'\u0944', u'\u0946',
                 u'\u0947', u'\u0948', u'\u0949', u'\u094A', u'\u094B', u'\u094C',
                 u'\u094d']
        limiters = ['.', '`', '!', ';', ', ', '?']

        virama = signs[-1] 
        lst_chars = []
        for char in word:
            if char in limiters:
                lst_chars.append(char)
            elif char in signs:
                if lst_chars:
                    lst_chars[-1] = lst_chars[-1] + char
                else:
                    continue
            else:
                if len(lst_chars) == 0:
                    lst_chars.append(char)
                else:
                    if lst_chars[-1][-1] == virama:
                        lst_chars[-1] = lst_chars[-1] + char
                    else:
                        lst_chars.append(char)
        return lst_chars
```

```python
data.columns = ["word","root"]
```

```python
lemmatized_list = []
for i in range(len(data)):   
    word = data.iloc[i][0]
    true_root = data.iloc[i][1]
    pred_root = get_root_word(word)

    # Create a dictionary representing a row of data
    row_data = {'Word': word, 'True_Root': true_root, 'Pred_Root': pred_root}

    lemmatized_list.append(row_data)

result_df = pd.DataFrame(lemmatized_list)
result_df
```

```python
def calculate_accuracy(result_df):
    correct_predictions = (result_df['True_Root'] == result_df['Pred_Root']).sum()

    total = len(result_df)
    accuracy = correct_predictions / total
    print("Accuracy : " , accuracy)
calculate_accuracy(result_df)
```
